```python
import pickle

import warnings
warnings.filterwarnings('ignore')
```

```python
def switcher(no):
    if(no=='1'):
       input1 = int(input("Enter Pregnancies(Number of times a woman has been pregnant) : "))
       input2= float(input("Enter Glucose :- Plasma Glucose concentration of 2 hours in an oral glucose tolerance test : "))
       input3 = float(input("Enter BloodPressure :- Diastollic Blood Pressure (mm hg) : "))
       input4 = int(input("Enter SkinThickness :- Triceps skin fold thickness(mm): "))  
       input5 = float(input("Enter Insulin :- 2 hour serum insulin(mu U/ml) : "))
       input6 = float(input("Enter BMI :- Body Mass Index ((weight in kg/height in m)^2) : "))
       input7 = float(input("Enter DiabetesPedigreeFunction :-scores likelihood of diabetes based on family history): "))
       input8 = int(input("Enter Age :- Age(years) : "))
       with open('diabetes_model_pkl' , 'rb') as f:
          lr = pickle.load(f)
       prediction = lr.predict([[input1,input2,input3,input4,input5,input6,input7,input8]])
       if prediction[0]==0:
         print("No diabetes")
       if prediction[0]==1:
         print("Diabetes Detected")
    elif(no=='2'):
       input1 = int(input("Enter Age : "))
       Sex = int(input("Enter Sex (1 - male , 0- female) : "))
       input3 = int(input("Enter Chest Pain Type(1,2,3,4) : "))
       input4 = int(input("Enter BP : "))  
       input5 = int(input("Enter chlolesterol : "))
       input6 = int(input("Enter FBS over 120 : "))
       input7 = int(input("Enter EKG : "))
       input8 = int(input("Enter Max HR : "))
       input9 = int(input("Enter Exercise Angia : "))
       input10 = float(input("Enter ST Depression : "))
       input11 = int(input("Enter Slope of ST : "))
       input12 = int(input("Enter No. of vessels fluoro : "))
       input13 = int(input("Enter Thallium : "))
       with open('heart_model_pkl' , 'rb') as f:
          lr = pickle.load(f)
       prediction = lr.predict([[input1,Sex,input3,input4,input5,input6,input7,input8,input9,input10,input11,input12,input13]])
       if prediction[0]==0:
         print("No Heart disease")
       if prediction[0]==1:
         print("Heart Disease Detected")
    elif(no=='3'):
       input1 = int(input("Enter Age : "))
       input2 = int(input("Enter Gender Female-0 ,Male-1: "))
       input3 = int(input("Enter Family History No-0 , Yes-1 : "))
       input4 = int(input("Enter benefits: Does your employer provide mental health benefits? : "))  
       input5 = int(input("Enter care_options: Do you know the options for mental health care your employer provides? : "))
       input6 = int(input("Enter anonymity: Is your anonymity protected if you choose to take advantage of mental health or substance abuse treatment resources? : "))
       input7 = int(input("Enter leave: How easy is it for you to take medical leave for a mental health condition? : "))
       input8 = int(input("Enter work_interfere: If you have a mental health condition, do you feel that it interferes with your work? : "))
       with open('mentalhealth_model_pkl' , 'rb') as f:
          lr = pickle.load(f)
       prediction = lr.predict([[input1,input2,input3,input4,input5,input6,input7,input8]])
       if prediction[0]==0:
         print("No Mental Health illness ")
       if prediction[0]==1:
         print("It seems your mental health is not so good")    
    elif(no=='4'):
       input1 = int(input("Enter Sample code number : "))
       input2 = int(input("Enter Clump Thickness  : "))
       input3 = int(input("Enter Uniformity of Cell Size : "))
       input4 = int(input("Enter Uniformity of Cell Shape : "))  
       input5 = int(input("Enter Marginal Adhesion : "))
       input6 = int(input("Enter Single Epithelial Cell Size : "))
       input7 = int(input("Enter Bare Nuclei : "))
       input8 = int(input("Enter Bland Chromatin : "))
       input9 = int(input("Enter Normal Nucleoli: "))
       input10 = int(input("Enter Mitoses: "))
       with open('breast_model_pkl' , 'rb') as f:
          lr = pickle.load(f)
       prediction = lr.predict([[input1,input2,input3,input4,input5,input6,input7,input8,input9,input10]])
       if prediction[0]==2:
         print("Benign")
       if prediction[0]==4:
         print("Malignant")
    elif(no=='Q'):
        print("Quiting...")
    else:
        print("Give input of numbers from 1 to 5")
```

```python
no = '0'
while( no != 'Q') :
  print('***************************************************\n')
  print("\t MAIN MENU. \n\n Please select below option \n 1. Diabetes Prediction\n 2. Heart Disease Prediction \n 3. Prediction of mental health illness \n 4. Breast Cancer Prediction \n  5. Quit(Press 'Q') \n")
  print('***************************************************\n')
  no=input('Enter a option ')
  switcher(no)
```
